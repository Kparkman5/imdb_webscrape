```python
# import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# assign global variables
URL = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'

header = {"User-Agent": "Mozilla/5.0 (X11; CrOS x86_64 12871.102.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.141 Safari/537.36"}

page = requests.get(URL, headers=header)
page.raise_for_status()

soup = BeautifulSoup(page.text, "html.parser")
movieList = soup.findAll('li', {'class': 'ipc-metadata-list-summary-item sc-3f724978-0 enKyEL cli-parent'})
final_dict = {'Rank': [], 'Title': [], 'imdb_rating': [], 'num_ratings': [], 'release_year': [], 'movie_rating': [], 'runtime': [], 'Director': [], 'Writers': [], 'Stars': [], 'num_user_reviews': [], 'num_critic_reviews': []}
```

```python
# scrape webpage for URL
def get_soup(URL):
    page = requests.get(URL, headers=header)
    page.raise_for_status()
    soup = BeautifulSoup(page.text, "html.parser")
    return soup
```

```python
# Find and assign all attributes to variables
def assign_values(soup):
    # Find and assign all attributes to variables
    rank = key.split(".")[0]
    title = key.split(".")[1].lstrip()
    imdb_rating = soup.find('span', {'class': 'sc-bde20123-1 cMEQkK'}).text
    ratings_count = soup.find('div', {'class': 'sc-bde20123-3 gPVQxL'}).text
    release_year = soup.find('ul', {'class': 'ipc-inline-list ipc-inline-list--show-dividers sc-d8941411-2 cdJsTz baseAlt'}).findChildren()[0].text
    movie_rating = soup.find('ul', {'class': 'ipc-inline-list ipc-inline-list--show-dividers sc-d8941411-2 cdJsTz baseAlt'}).findChildren()[2].text
    runtime = soup.find('ul', {'class': 'ipc-inline-list ipc-inline-list--show-dividers sc-d8941411-2 cdJsTz baseAlt'}).findChildren()[4].text
    director = soup.find('ul', {'class': 'ipc-inline-list ipc-inline-list--show-dividers ipc-inline-list--inline ipc-metadata-list-item__list-content baseAlt'}).findChildren()[0].text
    writers = soup.findAll('ul', {'class': 'ipc-inline-list ipc-inline-list--show-dividers ipc-inline-list--inline ipc-metadata-list-item__list-content baseAlt'})[1].findAll('li')
    stars = soup.findAll('ul', {'class': 'ipc-inline-list ipc-inline-list--show-dividers ipc-inline-list--inline ipc-metadata-list-item__list-content baseAlt'})[2].findAll('li')

    # Add attributes in final dictionary
    final_dict['Rank'].append(rank)
    final_dict['Title'].append(title)
    final_dict['imdb_rating'].append(imdb_rating)
    final_dict['num_ratings'].append(ratings_count)
    final_dict['release_year'].append(release_year)
    final_dict['movie_rating'].append(movie_rating)
    final_dict['runtime'].append(runtime)
    final_dict['Director'].append(director)
    wl = []
    for li in writers:
        wl.append(li.text)
        # print(li.text)
    final_dict['Writers'].append(wl)
    sl = []
    for li in stars:
        sl.append(li.text)
        # print(li.text)
    final_dict['Stars'].append(sl)
```

```python
# create list of URLs linked to each movie title
link_dict = {}
for i in movieList:
    a = i.find('a')
    title = i.find('h3').text
    # print(title)
    URL = 'https://www.imdb.com' + a['href']
    link_dict[title] = URL
    # print(URL)
del link_dict['77. Das Boot']
```

```python
for key in link_dict:
    soup = get_soup(link_dict[key])

    assign_values(soup)
    print(key.split(".")[0])

print(final_dict)
```

```python
df = pd.DataFrame(final_dict, columns=['Rank', 'Title', 'release_year', 'imdb_rating', 'num_ratings', 'movie_rating', 'runtime', 'Director', 'Writers', 'Stars'])
```
